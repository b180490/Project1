# Model Development
## Declaring dependent and independent variable

```python
X = df.drop(['price','transmission','state','title_status','Unnamed: 0','paint_color','type','condition','model'], axis=1)

Y = df['price']
```

### Training and Testing Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y , train_size=0.7, random_state=12345)
```

# Linear Regression Model

```python
def build_model_LR():
    from sklearn.linear_model import LinearRegression
    
    model = LinearRegression()
    
    model.fit(x_train,y_train)
    
    return model
```

# DecisionTree Model

```python
def build_model_DT():
    from sklearn.tree import DecisionTreeRegressor
    
    model = DecisionTreeRegressor()
    
    model.fit(x_train,y_train)

    with open('RandomForest003.pkl', 'wb') as file:
        pickle.dump(model, file)
    
    return model
```

# RandomForest Model

```python
def build_model_RF():
    from sklearn.ensemble import RandomForestRegressor
    
    model = RandomForestRegressor()
    
    model.fit(x_train,y_train)

    # with open('RandomForest001.pkl', 'wb') as file:
    #     pickle.dump(model, file)


    return model
```

# KNN Model

```python
def build_model_KNN():
    from sklearn.neighbors import KNeighborsRegressor
    
    model = KNeighborsRegressor()
    
    model.fit(x_train,y_train)
    
    return model
```

# xgboost Model

```python
def build_model_XGB():
    from xgboost import XGBRegressor
    
    model = XGBRegressor()
    
    model.fit(x_train,y_train)
    
    return model
```

# CatBoost Model

```python
def build_model_CTB():
    from catboost import CatBoostRegressor
    
    model = CatBoostRegressor()
    
    model.fit(x_train,y_train)

    with open('RandomForest002.pkl', 'wb') as file:
        pickle.dump(model, file)
    
    return model
```

```python
model_LR = build_model_LR()

model_DT = build_model_DT()

model_RF = build_model_RF()

model_KNN = build_model_KNN()

#model_XGB = build_model_XGB()

model_CTB = build_model_CTB()
```

```python
def evaluate_model(model):

    y_true = y_test
    y_pred = model.predict(x_test)
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    mae = mean_absolute_error(y_true,y_pred)
    mse = mean_squared_error(y_true,y_pred)
    r2 = r2_score(y_true,y_pred)

    print(f"Mean absolute error = {mae}")
    print(f"Mean squared error  = {mse}")
    print(f"R2 score = {r2}")
```

```python
evaluate_model(model_LR)
```

```python
evaluate_model(model_DT)
```

```python
x = df.drop(['price','transmission','state','title_status','Unnamed: 0'], axis=1)

y = df['price']
```

```python
# split x and y into training and testing dataset

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=12345)
```

```python
## MODEL BUILDING
```

```python
def build_dt_model():
    from sklearn.tree import DecisionTreeClassifier
    
    # create a model
    model = DecisionTreeClassifier()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_lg_model():
    from sklearn.linear_model import LogisticRegressionCV
    
    # create a model
    model = LogisticRegressionCV()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_knn_model():
    from sklearn.neighbors import KNeighborsClassifier
    
    # create a model
    model = KNeighborsClassifier(n_neighbors=9)
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_nb_model():
    from sklearn.naive_bayes import GaussianNB
    
    # create a model
    model = GaussianNB()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_svm_model():
    from sklearn.svm import SVC
    
    # create a model
    model = SVC()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def build_rf_model():
    from sklearn.ensemble import RandomForestClassifier
    
    # create a model
    model = RandomForestClassifier(n_estimators=100)
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```
